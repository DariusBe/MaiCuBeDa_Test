```python
from matplotlib import pyplot as plt

def display_image(image):
    plt.imshow(image, cmap='gray')
    plt.show()
```

```python
import cv2
import numpy as np
def conv(image, number_of_filters=1):
    #gray_image = np.reshape(gray_image, (gray_image.shape[0], gray_image.shape[1], 1))

    # create a kernel
    kernel = np.array([[-1, -1, -1],
                        [-1,  8, -1],
                        [-1, -1, -1]])
    
    # Apply convolution operation
    image = cv2.filter2D(image, -1, kernel)

    # if number_of_filters > 1 repeat the operation
    if number_of_filters > 1:
        for i in range(number_of_filters - 1):
            image = cv2.filter2D(image, -1, kernel)
            
    # Apply ReLU activation
    image = np.maximum(image, 0)
    return image
```

```python
import cv2
import numpy as np

# Load the image
image = cv2.imread(image_path)

# If the image is not loaded successfully, imread returns None
if image is None:
    print(f"Failed to load image at {image_path}")
else:
    print(f"Image loaded successfully from {image_path}")
    # convert image to BW
    image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    test = conv(image, number_of_filters=5)
    display_image(test)
```
