```python
import os

image_name = "line_1_HS_284_03_front.png" #"line_1_HS_1067_03_front.png"

parent_dir = os.path.abspath(os.path.join(os.getcwd(), os.pardir))
folder_path = os.path.join(parent_dir, "Cuneiform Tests", "MaiCuBeDa", "MaiCuBeDa_Annotations_Mixed","line")

image_path = folder_path = os.path.join(folder_path, image_name)

print(image_path)
```

```python
from matplotlib import pyplot as plt

def display_image(image):
    plt.imshow(image)
    plt.show()
```

```python
import numpy as np

def detect_contours(image):
    # common filtering techniques in cv2 are: thresholding, blurring, dilation, erosion, opening, closing, morphological transformations
    # other filters are kirsh, sobel, laplacian, canny edge detection
    # opening:
    image = cv2.morphologyEx(image.copy(), cv2.MORPH_OPEN, np.ones((9, 9), np.uint16))
    # kirsh:
    image = cv2.filter2D(image.copy(), -1, np.array([[-2, 4, -2], [-1, -5, -3], [3, 1, 6]]))
    # blurring:
    image = cv2.GaussianBlur(image.copy(), (5, 5), 0)
    # threshold the image
    _, image = cv2.threshold(image.copy(), 100, 255, cv2.THRESH_BINARY)
    # prepare image to be shown as non-numerical
    return image
```

```python
import cv2

# Load the image
image = cv2.imread(image_path)

# If the image is not loaded successfully, imread returns None
if image is None:
    print(f"Failed to load image at {image_path}")
else:
    print(f"Image loaded successfully from {image_path}")
    display_image(image)
    thresh = detect_contours(image)
    display_image(thresh)
```
